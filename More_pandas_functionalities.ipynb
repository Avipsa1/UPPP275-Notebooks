## Handling missing data in a `DataFrame`

```python
tracts.iloc[:,5][2] #Pick the 2nd item from the 6th column
```

```python
from numpy.core.numeric import NaN
tracts.iloc[:,5][2] = NaN #Introduce a missing values
```

```python
tracts.head()
```

```python
#drop any rows that have missing data.
tracts1 = tracts
tracts1.dropna(how = "any")
```

```python
#Fill missing data with a new value
tracts2 = tracts
tracts2.fillna(value = '999999')
```

### Merge:  Join two dataframes like we do in a database join. A common column in both dataframes will be used to tie together all columns into a single dataframe.

```python
crimes = pd.read_csv('CA-crimes_census_tract.csv')
crimes.head()
```

```python
merged_df = pd.merge(tracts, crimes, on='GEOID10')
merged_df.head()
```

## Groupby : Let us group values by categories using columns that have a non-numeric data type like Strings

```python
grouped_df = merged_df.groupby(by = 'county_name').mean()
grouped_df.head()
```

Now select only median home value, crime rates and total population from the grouped_df

```python
grouped_df[['total_pop','crime_rates','med_home_value']]
```

Now let us try to calculate correlations between the different columns

```python
grouped_df[['total_pop','crime_rates','med_home_value']].corr()
```
