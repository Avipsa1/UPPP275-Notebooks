### Merge:  Join two dataframes like we do in a database join. A common column in both dataframes will be used to tie together all columns into a single dataframe.

```python
crimes = pd.read_csv('CA-crimes_census_tract.csv')
crimes.head()
```

```python
tracts = pd.read_csv('CA_census_tracts.csv')
tracts.head()
```

```python
merged_df = pd.merge(tracts, crimes, on='GEOID10')
merged_df.head()
```

## Groupby : Let us group values by categories using columns that have a non-numeric data type like Strings

```python
grouped_df = merged_df.groupby(by = 'county_name').mean()
grouped_df.head()
```

Now select only median home value, crime rates and total population from the grouped_df

```python
grouped = grouped_df[['total_pop','crime_rates','med_home_value']]
grouped.shape
```

```python
grouped.head()
```

Now let us try to calculate correlations between the different columns

```python
grouped[['total_pop','crime_rates']].corr()
```

## Plotting functions in pandas

```python
grouped['county'] = grouped.index
grouped.head()
```

```python
#Select only the columns 'county' and 'crime_rates' from grouped
df = grouped[['county','crime_rates']]
#Using the plot function create a barplot

df.plot(kind = 'bar', figsize = (10,10), color = 'purple')
```

```python
df.plot(kind = 'barh', figsize = (10,10), color = 'gold')
```

```python
#Create boxplot
df2 = merged_df[['pct_black',	'pct_hispanic',	'pct_white']]
df2.plot(kind = 'box', color = 'green', patch_artist = True, figsize = (10,10))
```
