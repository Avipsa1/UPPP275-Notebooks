Often times we need to apply a function to a column in a dataframe to transform it. 'pandas' makes it easier to operate on columns using the '.apply(') function. 

Let us try to map the values in the “geography_type” column to either a “1” or “0” depending on the original value. We will create a new column append this information to the DataFrame in a new column.

```python
#We write a new function that checks if the geography type is a 'City' and returns 1 if True and 0 if False
def mapGeography(x):
    if x == "City":
        return 1
    else:
        return 0
```

```python
#Now, lets add the values generated by this function to the 'new_column'
df['new_column'] = df.geography_type.apply(mapGeography)
df.head()
```

Let us count how many geography types are 'City' and how many are not using the value_counts() function

```python
df['new_column'].value_counts()
```

Let us now try to visualize the education levels in a barplot for all those geographies which are cities

```python
df_cities_edu_levels.index = df_cities_edu_levels.geography
```
