You can download the location of airports in the Southern California region here: https://gisdata-scag.opendata.arcgis.com/datasets/SCAG::airports-scag-region/explore?location=33.745253%2C-116.923250%2C8.94 

Download the GeoJSON file format and upload to your colab notebook worksapce.

```python
#Read geojson file
airports = gpd.read_file('Airports_SCAG_Region.geojson')
airports.head()
```

Now use the marker cluster function from folium like before but with the new geodataframe 'airports' to map the points.

```python
# Load map centred on Southern California
m = folium.Map(location=[33.66319,-117.83125], zoom_start=10, control_scale=True)
 
# create a marker cluster for airport locations in SCAG region
marker_cluster = folium.plugins.MarkerCluster().add_to(m)

#Loop over the x and y coordinates from the geometry column in the 'airports' geodataframe
for row in airports.itertuples():
    folium.Marker(location=[row.geometry.y,row.geometry.x],popup=row.NAME).add_to(marker_cluster)
```
